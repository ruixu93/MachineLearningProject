```python
import pandas as pd
import numpy as np
import time
```

```python
column_name = ['taxi_id', 'time_stamp', 'week_day', 'trip_trace']
```

# for all taxi, modify the dataset

```python
# get all the names of taxi
taxi_names = []
f = open("/home/rui/Documents/MLProject/data/crab/cabspottingdata/_cabs.txt", "r")
while True:
    line = f.readline()
    if line:
        line_split = line.split(" ")
        taxi_names.append(line_split[1].replace("id=","").replace("\"",""))
    else:
        break
f.close
```

```python
# initialize some lists to store data
taxi_id = []
time_stamp = []
week_day = []
trip_trace = []
sumtaix = 0

# loop for all taxis
for taxi_name in taxi_names:
    taxi_trace = pd.read_csv("/home/rui/Documents/MLProject/data/crab/cabspottingdata/new_" + taxi_name + ".txt", header=None, sep=" ")
    num_row = taxi_trace.shape[0]
    previous_occupied = 3
    sumtaix += 1
    print sumtaix
    
    for i in range(num_row):
        row = taxi_trace.loc[num_row-1-i,:]
        latitude = row[0]
        longitude = row[1]
        current_occupied = row[2]
        timestamp = row[3]
        
        #
        if previous_occupied == 3 and current_occupied == 0:
            print "from this"
            previous_occupied = current_occupied
            continue
            
        if previous_occupied == 3 and current_occupied == 1:
            print "skip this"
            continue
        
        # not begin a trip
        if current_occupied == 0 and previous_occupied == 0:
            previous_occupied = current_occupied
            continue

        # begin a trip, create a new list
        if current_occupied == 1 and previous_occupied == 0:
            new_trace = []
            new_trace.append([[latitude, longitude], timestamp])

        # trip continuing
        if current_occupied == 1 and previous_occupied == 1:
            new_trace.append([[latitude, longitude], timestamp])

        # end of a trip
        if current_occupied == 0 and previous_occupied == 1:
            if len(new_trace) > 1:
                trip_trace.append(new_trace)
                taxi_id.append(taxi_name)
                time_stamp.append(timestamp)
                week_day.append(time.gmtime(timestamp)[6])
        
        # end of a single taxi dataset
        if i == (num_row - 1) and current_occupied == 1:
            new_trace = []
            
        previous_occupied = current_occupied
        
        
```

```python
print len(taxi_id), len(time_stamp), len(week_day), len(trip_trace)
```

```python
taxi_pandas = pd.DataFrame({'taxi_id': taxi_id, 'time_stamp': time_stamp, 'week_day': week_day})
taxi_pandas['trip_trace'] = trip_trace
```
